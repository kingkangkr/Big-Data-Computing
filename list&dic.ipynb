```python
scores=["A+","B0","B+","A-","C+","A+","B0","C0"]

#받은 성적을 리스트로 표현 

dic={'A+':4.3,'A0':4.0,'A-':3.7,'B+':3.3,'B0':3.0,'B-':2.7,'C+':2.3,'C0':2.0,'C-':1.7,'D+':1.3,'D0':1.0,'D-':1.7}

#성적이 몇 점을 나타내는지 딕셔너리로 표현 

credit_hours=[3,3,1,2,2,3,3,3]

#이수 학점을 리스트로 표현 

num=0

#num에 0 할당 

total_score=0

#가중치를 매긴 성적들의 합에 0 할당

while(num<len(scores)):#num이 scores의 길이보다 작을 동안 

    total_score+=(credit_hours[num]*dic[scores[num]])

    #최종 성적은 가중치 * 받은 성적 
    
    num+=1#num 증가 

time=0

#가중치의 합에 0 할당 

num=0

#num에 0 할당 

while(num<len(credit_hours)):

#num이 가중치의 개수보다 작을 동안 

    time+=credit_hours[num]

    #가중치의 합 

    num+=1

    #num 증가 

total_score/time

#가중치 * 받은 성적의 총 합에서 가중치로 나누면 평균 성적을 구할 수 있다.

```

```python
f = open("score.txt", 'r')

#score.txt를 읽기 형식으로 연다.

scores=f.readlines()

#readlines 메서드를 이용해 score.txt의 내용을 scores에 저장한다.

total=[]

#total이라는 빈 리스트 생성 

ans=[]

#ans라는 빈 리스트 생성 

for score in scores:

#scores의 각 줄 인 score를 차례대로 읽어서 

    x=score.strip().split(',')

    #score를 ,를 기준으로 나눈 후, strip을 이용하여 공백을 없앤다.

    #여기서 x는 string이다. 

    inner_list=[]

    #inner_list라는 리스트 생성 

    for y in x:

    #x의 각 원소인 y를 

        inner_list.append(int(y.strip()))

        #inner_list에 int형식으로 변환한 후 추가 

    total.append(inner_list)

    #total에 추가하여서 리스트를 정수의 리스트로 바꿈 

for x in total:

#total의 각 리스트 값을 x에 배정 

    temp=0

    #temp에 0 배정 

    for y in x:

    #x의 각 원소인 y를 

        temp+=y

        #temp에 더해준다

    ans.append(temp/5)

    #temp를 5로 나눈 값을 ans리스트에 추가 

print(ans)

#평균의 list인 ans 출력 

f.close()

#f를 닫는다. 

```
