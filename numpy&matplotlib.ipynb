```python
import numpy as np
import csv
import matplotlib.pyplot as plt
f=open('./age_no_comma.csv')
#파일을 f로 open
data=csv.reader(f)
#f를 읽어서 csv reader로 읽어서 data에 저장.
next(data)
#data를 계속 읽는다.
data=list(data)
#다시 읽어오지 않기 위해 저장
max=0
#0세 이상, 6세 이하 비율이 최대인 곳의 비율 저장
city=''
#유아 비율 최대인곳의 이름 저장
for row in data:
    #for문을 이용해 모든row 탐색
    if(int(row[1])==0):
        #인구수가 0인곳은 스킵
        continue
    kid=0
    #0세부터 6세의 인구수 초기화
    for i in range(7):
        #0세부터 6세이니 range(7)
        kid+=int(row[i+3])
        #0세부터 6세의 인구수를 모두 더한다.
    if(max<(kid/int(row[1]))):
        #0세부터 6세의 인구의수 나누기 전체 인구수를 해서
        max=kid/int(row[1])
        #max 값보다 크면 max 값에 비율 저장
        city=row[0]
        #city에 도시 이름 저장
print(city+'지역의 0세 이상 6세 이하의 비율:'+str((max*100))+"%")
#비율 가장 높은곳의 이름과 비율 출력
for row in data:
    #for문을 이용해 row탐색
    if city in row[0]:
        #비율이 가장 높은 곳의 인구
        home=np.array(row[3:],dtype=int)
        #0세의 인구인 4번째 column부터 home에 추가
plt.plot(home)
#home을 plot
plt.show()
#결과 출력
```
